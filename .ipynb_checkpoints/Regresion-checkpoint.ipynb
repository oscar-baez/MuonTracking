```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input, Model
```

```python
#data = df_ccd1_clean[['X_C', 'Y_C']].values
data = df_ccd1_clean[['X_C', 'X_asymmetry','Y_C', 'Y_asymmetry']].values
predict = df_ccd2_clean[['X_C', 'X_asymmetry','Y_C', 'Y_asymmetry']].values
#data = df_ccd1_clean[['X_C', 'X_asymmetry']].values
#predict = df_ccd2_clean[['X_C', 'X_asymmetry']].values
```

```python
df_ccd1_clean.head()
```

```python
data
```

```python
# Build the model
model1 = Sequential()
model1.add(Dense(1, activation='linear' , input_shape=(2,), name = "Capa_de_Entrada"))
model1.compile(loss='mse', optimizer='adam')
```

```python
model1.summary()
```

```python
data_train, data_test, labels_train, labels_test = train_test_split(data, predict)
```

```python
history = model1.fit(data_train, labels_train, epochs=10000, validation_split=0.2)
```

```python
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```
