# Functions

```python
def centeroidnp(x_list,y_list):
    lengthx = len(x_list)
    lengthy = len(y_list)
    sum_x = np.sum(x_list)
    sum_y = np.sum(y_list)
    return sum_x/lengthx, sum_y/lengthy
```

```python
def getParameters(x_list,y_list,eventNumber):
    x_list = [x for x in x_list if not np.isnan(x)]
    y_list = [y for y in y_list if not np.isnan(y)]
    x_c ,y_c = np.mean(x_list), np.mean(y_list)
    x_min ,x_max= int(np.nanmin(x_list)), int(np.nanmax(x_list))
    y_min ,y_max= int(np.nanmin(y_list)), int(np.nanmax(y_list))
    length= len(x_list)
    x_asymmetry, y_asymmetry = (x_c - (x_max + x_min)/2), (y_c - ( y_max + y_min)/2)
    x_size,y_size = x_max-x_min, y_max-y_min
    #print("EventNumber ={} X_C={} Y_C={} X_Size={} Y_Size={} X_asymmetry={} Y_asymmetry={}".format(eventNumber,x_c,y_c,x_size,y_size,x_asymmetry,y_asymmetry))
    return [eventNumber,x_c,y_c,x_size,y_size,x_asymmetry,y_asymmetry]
```

```python
def getParameters_d(x_list,y_list,eventNumber):
    #print("***")
    #indices_sin_nan = np.where(~np.isnan(x_list))[0]
    #print(indices_sin_nan)
    #x_list = x_list[indices_sin_nan]
    #indices_sin_nan = list(np.where(~np.isnan(y_list)))
    #print(indices_sin_nan)[0]
    #y_list = y_list[indices_sin_nan]
    
    
    x_list = [x for x in x_list if not np.isnan(x)]
    y_list = [y for y in y_list if not np.isnan(y)]

    x_min ,x_max= int(np.nanmin(x_list)), int(np.nanmax(x_list))
    y_min ,y_max= int(np.nanmin(y_list)), int(np.nanmax(y_list))
    MPP=15
    binx, biny= int((x_max - x_min) / MPP),  int((y_max - y_min) / MPP)
    if binx==0 or biny ==0:
        return False
    #print(binx, biny)
    frecuenciasx, binsx = np.histogram(x_list, bins=binx) 
    frecuenciasy, binsy = np.histogram(y_list, bins=biny)
    x_max ,x_min, y_max ,y_min = max(binsx) ,min(binsx), max(binsy), min(binsy)

    x_c ,y_c = np.sum(binsx[:-1] * frecuenciasx) / np.sum(frecuenciasx) , np.sum(binsy[:-1] * frecuenciasy) / np.sum(frecuenciasy) 
    x_asymmetry, y_asymmetry = (x_c - (x_max + x_min)/2), (y_c - ( y_max + y_min)/2)
    x_size,y_size = x_max-x_min, y_max-y_min
    
    return [eventNumber,x_c,y_c,x_size,y_size,x_asymmetry,y_asymmetry]
```
